# 계절별 오염물질 분포

```python
# 측정일시 열을 datetime 형식으로 변환
air_21['측정일시'] = pd.to_datetime(air_21['측정일시'])

# 'Month'와 'Hour'를 정확하게 추출
air_21['Month'] = air_21['측정일시'].dt.month
air_21['Hour'] = air_21['측정일시'].dt.hour

# 계절 정보 함수 정의
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Season 열 설정
air_21['Season'] = air_21['Month'].apply(get_season)

# 시각화할 대기 오염 물질 리스트
pollutants = ['SO2', 'CO', 'O3', 'NO2', 'PM10', 'PM25']

# 계절별 시각화
plt.figure(figsize=(14, 10))
for i, pollutant in enumerate(pollutants, 1):
    plt.subplot(3, 2, i)
    for season in ['Spring', 'Summer', 'Fall', 'Winter']:
        seasonal_data = air_21[air_21['Season'] == season]
        sns.lineplot(
            x='Hour',
            y=pollutant,
            data=seasonal_data,
            estimator='mean',
            errorbar=None,
            label=season
        )
    plt.title(f'Hourly Change in {pollutant} by Season')
    plt.xlabel('Hour')
    plt.ylabel('Concentration')
    plt.legend(title='Season')
plt.tight_layout()
plt.show()
```

```python
# 시간 형식 확인 및 변환
try:
    air_21['시간'] = pd.to_datetime(air_21['시간'])
    air_21['Month'] = air_21['시간'].dt.month
    air_21['Hour'] = air_21['시간'].dt.hour
    air_21['Season'] = air_21['Month'].apply(get_season)
except Exception as e:
    print("Datetime conversion error:", e)

# 변환 결과 확인
print(air_21[['시간', 'Month', 'Hour', 'Season']].head())
plt.figure(figsize=(10, 6))
for i, pollutant in enumerate(pollutants, 1):
    plt.subplot(3, 2, i)
    sns.lineplot(x='시간', y=pollutant, data=air_22, estimator='mean', errorbar=None)
    plt.title(f'Hourly Change in {pollutant}')
    plt.xlabel('Hour')
    plt.ylabel('Concentration')
plt.tight_layout()
plt.show()
```

```python
# 시간 형식 확인 및 변환
try:
    air_21['시간'] = pd.to_datetime(air_21['시간'])
    air_21['Month'] = air_21['시간'].dt.month
    air_21['Hour'] = air_21['시간'].dt.hour
    air_21['Season'] = air_21['Month'].apply(get_season)
except Exception as e:
    print("Datetime conversion error:", e)

# 변환 결과 확인
print(air_21[['시간', 'Month', 'Hour', 'Season']].head())

# 배경색을 #181818로 설정
plt.figure(figsize=(10, 6), facecolor='#181818')  # 전체 배경색 설정

for i, pollutant in enumerate(pollutants, 1):
    plt.subplot(3, 2, i, facecolor='#181818')  # 각 서브플롯 배경색 설정
    sns.lineplot(x='시간', y=pollutant, data=air_22, estimator='mean', errorbar=None, color='#c89f65')  # 선 색상을 노란색으로 설정
    plt.title(f'Hourly Change in {pollutant}', color='white')  # 제목 색상 조정
    plt.xlabel('Hour', color='white')  # x축 라벨 색상 조정
    plt.ylabel('Concentration', color='white')  # y축 라벨 색상 조정
    plt.tick_params(axis='x', colors='white')  # x축 눈금 색상 조정
    plt.tick_params(axis='y', colors='white')  # y축 눈금 색상 조정

plt.tight_layout()
plt.show()
```
